```python
import pandas as pd
bnamesdf= pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv')
bnamesdf.head(10)
```

```python
#1. Delete unnamed columns
bnamesdf.drop(bnamesdf.columns[bnamesdf.columns.str.contains('unnamed',case = False)],axis = 1,inplace=True)
bnamesdf.columns
```

```python
#2. Show the distribution of male and female
bnamesdf.groupby("Gender").describe()
```

```python
#3. Show the top 5 most preferred names
bnamesdf["Name"].value_counts().nlargest(5)
```

```python
#4. the median name occurence in the dataset
import math
bnamesdf.sort_values("Name", inplace=True)
middle = math.ceil(len(bnamesdf)/2)
bnamesdf.iloc[middle,1]
```

```python
#5. Distribution of male and female born count by states
bnamesdf.groupby(["State","Gender"]).count()
```
