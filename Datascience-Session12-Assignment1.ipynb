```python
#Read the following data set:https://archive.ics.uci.edu/ml/machine-learning-databases/adult/
#Rename the columns as per the description from this file:
#https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.names
import pandas as pd
colnames=["age","workclass","fnlwgt","education","education_num","marital_status","occupation",
                       "relationship","race","sex","capital_gain",
                        "capital_loss","hours_per_week","native_country","prediction"]
adultdf=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data",names=colnames)
adultdf.columns

```

```python
#1. Select 10 records from the adult sqladb
from pandasql import sqldf
sqladb = lambda q: sqldf(q, globals())
q = """ SELECT * FROM adultdf LIMIT 10; """
sqladb(q)
```

```python
#Show me the average hours per week of all men who are working in private sector
q = """
    SELECT 
       avg(hours_per_week) as hoursperweek
    FROM 
        adultdf a1 
     WHERE a1.workclass = " Private" """
sqladb(q)
```

```python
#Show me the frequency table for education, occupation and relationship, separately
q = """
     SELECT  a1.education, COUNT(*) as Frequency FROM adultdf a1 group by a1.education """

sqladb(q)
```

```python
#Show me the frequency table for education, occupation and relationship, separately
q = """
     SELECT  a1.occupation, COUNT(*) as Frequency FROM adultdf a1 group by a1.occupation """

sqladb(q)
```

```python
#Show me the frequency table for education, occupation and relationship, separately
q = """
     SELECT  a1.relationship, COUNT(*) as Frequency FROM adultdf a1 group by a1.relationship """

sqladb(q)
```

```python
#Are there any people who are married, working in private sector and having a masters degree
q = """
     SELECT  count(*) as PeopleCount FROM adultdf a1 where a1.marital_status like " Married%" and a1.workclass = " Private" and a1.education=" Masters" """
#
df=sqladb(q)
#if ([["PeopleCount"]] >0):
 #   
print(df[["PeopleCount"]])
print("Yes. There are People who are married, working in private sector and having a masters degree ")
```

```python
#What is the average, minimum and maximum age group for people working in different sectors
q = """
    SELECT 
       workclass,
       avg(age) as average,
       min(age) as Minimum,
       max(age) as Maximum      
    FROM 
        adultdf a1 
    group by a1.workclass """
sqladb(q)
```

```python
#Calculate age distribution by country
q = """
    SELECT 
       nativecountry,
       avg(age) as average,
       min(age) as Minimum,
       max(age) as Maximum      
    FROM 
        adultdf a1 
    group by a1.nativecountry """
sqladb(q)
```

```python
#Compute a new column as 'Net-Capital-Gain' from the two columns 'capital-gain' and 'capital-loss'
q = """
    SELECT 
       (capital_gain-capital_loss) as 'Net-Capital-Gain'      
    FROM 
        adultdf a1 
    """
sqladb(q)
```
